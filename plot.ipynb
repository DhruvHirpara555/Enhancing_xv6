```python
# read from test.txt

import matplotlib.pyplot as plt
import numpy as np

# open file
f = open('test.txt', 'r')

# p4 = [(int(y),int(x)) for line in f.readlines() for x,y,z in line.split() if z == '4']
# p5 = [(y,x) for line in f.readlines() for x,y,z in line.split() if z == '5']
# p6 = [(y,x) for line in f.readlines() for x,y,z in line.split() if z == '6']
# p7 = [(y,x) for line in f.readlines() for x,y,z in line.split() if z == '7']
# p8 = [(y,x) for line in f.readlines() for x,y,z in line.split() if z == '8']
# p9 = [(y,x) for line in f.readlines() for x,y,z in line.split() if z == '9']

p4x = []
p4y = []
p5x = []
p5y = []
p6x = []
p6y = []
p7x = []
p7y = []
p8x = []
p8y = []
p9x = []
p9y = []

for i in range(4, 10):
    for line in f:
        if line[0] == str(i):
            # parse line
            line = line.split()
            if i == 4:
                # p4.append(((float(line[2])), (float(line[1]))))
                p4x.append(float(line[2]))
                p4y.append(float(line[1]))
            elif i == 5:
                # p5.append(((float(line[1])), (float(line[2]))))
                p5x.append(float(line[2]))
                p5y.append(float(line[1]))
            elif i == 6:
                # p6.append(((float(line[1])), (float(line[2]))))
                p6x.append(float(line[2]))
                p6y.append(float(line[1]))
            elif i == 7:
                # p7.append(((float(line[1])), (float(line[2]))))
                p7x.append(float(line[2]))
                p7y.append(float(line[1]))
            elif i == 8:
                # p8.append(((float(line[1])), (float(line[2]))))
                p8x.append(float(line[2]))
                p8y.append(float(line[1]))
            elif i == 9:
                # p9.append(((float(line[1])), (float(line[2]))))
                p9x.append(float(line[2]))
                p9y.append(float(line[1]))
        else:
            continue
            
    f.seek(0)

# close file
f.close()

# plot
# plt.plot(p4[:][0], p4[:][1], 'ro')
# plt.plot(p4x,p4y)
plt.title('MLFQ Aging Ticks = 30')
plt.xlabel('Ticks')
plt.ylabel('Current Queue')
plt.plot(p5x,p5y)
plt.plot(p6x,p6y)
plt.plot(p7x,p7y)
plt.plot(p8x,p8y)
plt.plot(p9x,p9y)
# increase plot size
plt.rcParams["figure.figsize"] = (25,10)
plt.legend(['pid = 5', 'pid = 6', 'pid = 7', 'pid = 8', 'pid = 9'], loc='upper left')
plt.show()

```
